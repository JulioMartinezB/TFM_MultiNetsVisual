```python
import pandas as pd
import numpy as np
```

```python
tramos_bus = pd.read_csv('Tramos_Bus.csv',header=0)
tramos_bus_interurbanos = pd.read_csv('Data_Interurbanos/trips.txt',header=0)
intercambiadores = pd.read_csv('Intercambiadores.csv',header=0)
```

```python
print(tramos_bus.columns)
tramos_bus.head()
```

```python
print(tramos_bus_interurbanos.columns)
tramos_bus_interurbanos['NUMEROLINEAUSUARIO'] = tramos_bus_interurbanos['route_id'].str.extract('8__(\d+)__')
tramos_bus_interurbanos.head()
```

```python
lineas_bus_necesarias = intercambiadores[intercambiadores['TipoTransporte']=='EMT']['Linea'].unique()
lineas_bus_interurbanas_necesarias = intercambiadores[intercambiadores['TipoTransporte']=='Interurbano']['Linea'].unique()
```

```python
tramos_bus_necesarios = tramos_bus[tramos_bus['NUMEROLINEAUSUARIO'].isin(lineas_bus_necesarias)]
tramos_bus_necesarios.head(2)
```

```python
tramos_bus_interurbanos_necesarios = tramos_bus_interurbanos[tramos_bus_interurbanos['NUMEROLINEAUSUARIO'].isin(lineas_bus_interurbanas_necesarias)]
tramos_bus_interurbanos_necesarios.head(2)
```

```python
ey = tramos_bus_necesarios[tramos_bus_necesarios['NUMEROLINEAUSUARIO']=='45']
#drop na columns
ey = ey.dropna(axis=1)
ey=ey.drop(columns=['MODOLINEA','LONGITUDTRAMOANTERIOR','FECHAALTA','FECHAINICIO','VELOCIDADTRAMOANTERIOR','CODIGOSUBLINEA'])
ey=ey.drop(columns=['CODIGOGESTIONLINEA','NUMEROLINEAUSUARIO','MODO','FECHAACTUAL','CODIGOPROVINCIA','CODIGOMUNICIPIO','MUNICIPIO','CORONATARIFARIA'])
ey[ey['SENTIDO']==1]
```

```python
paradas_bus = pd.read_csv('Paradas_Bus.csv',header=0)
paradas_bus_necesarias = paradas_bus[paradas_bus['LINEAS'].any().isin(lineas_bus_necesarias)]
paradas_bus
```
